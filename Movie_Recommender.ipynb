```python
import pandas as pd
import numpy as np
import streamlit as st
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
##importing and cleaning data for use
movies = pd.read_csv('/content/movies2024.csv')
movies = movies[['Title', 'Genres']]
movies.dropna(inplace = True)
movies
```

```python
##vectorization of movies
cv = CountVectorizer()
vectors = cv.fit_transform(movies['Genres'])
```

```python
##cosine similarity
similarity = cosine_similarity(vectors)
```

```python
##sacing the files for instant reading to avoid repeated calculations
pickle.dump(similarity, open('similarity.pkl', 'wb'))
pickle.dump(movies, open('movies_dict.pkl', 'wb'))

## for read back
similarity = pickle.load(open('similarity.pkl', 'rb'))
movies = pickle.load(open('movies_dict.pkl', 'rb'))
```

```python
##recommendation fucntion
def recommend(movie_index):
  score = similarity[movie_index]
  top5 = sorted(enumerate(score), reverse = True, key = lambda x:x[1])[1:6]
  return top5
```

```python
import streamlit as st
import pickle

# Load the pre-processed data and similarity matrix
similarity = pickle.load(open('similarity.pkl', 'rb'))
movies = pickle.load(open('movies_dict.pkl', 'rb'))

def recommend(movie_title):
  # Get movie index from title
  movie_index = movies[movies['Title'] == movie_title].index[0]
  score = similarity[movie_index]

  # Sort movies by similarity score and get top 5 (excluding itself)
  top5_indices = sorted(enumerate(score), reverse=True, key=lambda x: x[1])[1:6]

  # Get movie titles for the top 5 recommendations
  recommended_movies = []
  for i in top5_indices:
    recommended_movies.append(movies.iloc[i[0]].Title)
  return recommended_movies


st.title('Movie Recommender System')
option = st.selectbox('Pick a Movie', movies['Title'].values)

if st.button('Recommend'):
  names = recommend(option)

  col1, col2, col3, col4, col5 = st.columns(5)

  with col1:
    st.header(names[0])
  with col2:
    st.header(names[1])
  with col3:
    st.header(names[2])
  with col4:
    st.header(names[3])
  with col5:
    st.header(names[4])

# To save this code as app.py for streamlit deployment
with open('app.py', 'w') as f:
    f.write('''
import streamlit as st
import pickle
import pandas as pd

# Load the pre-processed data and similarity matrix
similarity = pickle.load(open('similarity.pkl', 'rb'))
movies = pickle.load(open('movies_dict.pkl', 'rb'))

def recommend(movie_title):
  # Get movie index from title
  movie_index = movies[movies['Title'] == movie_title].index[0]
  score = similarity[movie_index]

  # Sort movies by similarity score and get top 5 (excluding itself)
  top5_indices = sorted(enumerate(score), reverse=True, key=lambda x: x[1])[1:6]

  # Get movie titles for the top 5 recommendations
  recommended_movies = []
  for i in top5_indices:
    recommended_movies.append(movies.iloc[i[0]].Title)
  return recommended_movies


st.title('Movie Recommender System')
option = st.selectbox('Pick a Movie', movies['Title'].values)

if st.button('Recommend'):
  names = recommend(option)

  col1, col2, col3, col4, col5 = st.columns(5)

  with col1:
    st.header(names[0])
  with col2:
    st.header(names[1])
  with col3:
    st.header(names[2])
  with col4:
    st.header(names[3])
  with col5:
    st.header(names[4])
''')

print("Streamlit app code saved to app.py")
```
